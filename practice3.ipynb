## 一、for 循环


### 练习 01 · 从 1 到 100 的整数求和
**Version:** 1.0 · **知识点:** for 循环、range

**预期输出:** `5050`


```python
total = 0
for i in range(1, 100):
    total += i
print(total)
```

**Version:** 1.0 · **知识点:** for 循环、条件判断

## 二、while 循环


### 练习 05 · 从 1 到 100 的整数求和
**Version:** 1.1 · **知识点:** while 循环

**预期输出:** `5050`


```python
total = 0
i = 1
while i <= 100:
    total += i
    i += 1
print(total)
```

### 练习 06 · 从 1 到 100 的偶数求和
**Version:** 1.3 · **知识点:** while 循环

**预期输出:** `2550`


```python
total = 0
i = 2
while i <= 100:
    total += i
    i += 2
print(total)
```

## 三、break & continue


### 练习 07 · 从 1 到 100 的偶数求和
**Version:** 1.4 · **知识点:** while True、break

**预期输出:** `2550`


```python
total = 0
i = 2
while True:
    total += i
    i += 2
    if i > 100:
        break
print(total)
```

### 练习 08 · 从 1 到 100 的偶数求和
**Version:** 1.5 · **知识点:** continue

**预期输出:** `2550`


```python
total = 0
for i in range(1, 101):
    if i % 2 != 0:
        continue
    total += i
print(total)
```

## 四、嵌套循环


### 练习 09 · 打印乘法口诀表
**Version:** 1.0 · **知识点:** 嵌套 for 循环

**预期输出:** 9 行乘法口诀表


```python
for i in range(1, 10):
    for j in range(1, i + 1):
        print(f'{i} x {j} = {i * j}', end='\t')
    print()
```

### 练习 11 · 最大公约数（枚举法）
**Version:** 1.0 · **知识点:** for 循环、break

**演示数据:** `a=12, b=8` · **预期输出:** `a和b的最大公约数为4`


```python
a, b = 12, 8
small = min(a, b)
for i in range(1, small + 1):
    if a % i == 0 and b % i == 0:
        print(f'a和b的最大公约数为{i}')
        break
```

### 练习 12 · 最大公约数（辗转相除法）
**Version:** 1.1 · **知识点:** while 循环

**演示数据:** `a=12, b=8` · **预期输出:** `最大公约数为4`


```python
a, b = 12, 8
while a % b != 0:
    a, b = b, a % b
print(f'最大公约数为{b}')
```

### 练习 13 · 猜数字小游戏
**Version:** 1.0 · **知识点:** random、while 循环

> 需交互运行：单独执行此 cell，按提示输入猜测数字。


```python
import random

answer = random.randint(1, 100)
count = 0
print('猜字游戏开始‼️')
while True:
    guess = int(input("请输入数字："))
    count += 1
    if answer > guess:
        print('猜小了')
    elif answer < guess:
        print('猜大了')
    else:
        print('恭喜猜中了🎉')
        break
print(f'您一共猜了{count}次')
```
