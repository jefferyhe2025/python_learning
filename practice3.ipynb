## 一、for 循环


### 练习 01 · 从 1 到 100 的整数求和
**Version:** 1.0 · **知识点:** for 循环、range

**预期输出:** `5050`


```python
total_ex01 = 0
for i in range(1, 101):
    total_ex01 += i
print(total_ex01)
```

### 练习 02 · 从 1 到 100 的偶数求和
**Version:** 1.0 · **知识点:** for 循环、条件判断

**预期输出:** `2550`


```python
total_ex02 = 0
for i in range(100):
    if i % 2 == 0:
        total_ex02 += i
print(total_ex02)
```

### 练习 03 · 从 1 到 100 的偶数求和
**Version:** 1.1 · **知识点:** range 步长

**预期输出:** `2550`


```python
total_ex03 = 0
for i in range(2, 100, 2):
    total_ex03 += i
print(total_ex03)
```

### 练习 04 · 从 1 到 100 的偶数求和
**Version:** 1.2 · **知识点:** 内置 sum、range

**预期输出:** `2550`


```python
print(sum(range(2, 100, 2)))
```

## 二、while 循环


### 练习 05 · 从 1 到 100 的整数求和
**Version:** 1.1 · **知识点:** while 循环

**预期输出:** `5050`


```python
num_ex05 = 0
while num_ex05 < 100:
    num_ex05 += 1
total_ex05 = sum(range(1, num_ex05 + 1))
print(total_ex05)
```

### 练习 06 · 从 1 到 100 的偶数求和
**Version:** 1.3 · **知识点:** while 循环

**预期输出:** `2550`


```python
total_ex06 = 0
i_ex06 = 2
while i_ex06 <= 100:
    total_ex06 += i_ex06
    i_ex06 += 2
print(total_ex06)
```

## 三、break & continue


### 练习 07 · 从 1 到 100 的偶数求和
**Version:** 1.4 · **知识点:** while True、break

**预期输出:** `2550`


```python
total_ex07 = 0
i_ex07 = 2
while True:
    total_ex07 += i_ex07
    i_ex07 += 2
    if i_ex07 > 100:
        break
print(total_ex07)
```

### 练习 08 · 从 1 到 100 的偶数求和
**Version:** 1.5 · **知识点:** continue

**预期输出:** `2550`


```python
total_ex08 = 0
for i in range(1, 101):
    if i % 2 != 0:
        continue
    total_ex08 += i
print(total_ex08)
```

## 四、嵌套循环


### 练习 09 · 打印乘法口诀表
**Version:** 1.0 · **知识点:** 嵌套 for 循环

**预期输出:** 9 行乘法口诀表


```python
for i in range(1, 10):
    for j in range(1, i + 1):
        print(f'{i} x {j} = {i * j}', end='\t')
    print()
```

## 五、循环的应用


### 练习 10 · 判断素数
**Version:** 1.0 · **知识点:** 循环、break

**建议测试输入:** `17`（素数）、`18`（合数）

> 运行此 cell 后按提示输入正整数。


```python
num_ex10 = int(input("请输入一个正整数："))
end_ex10 = int(num_ex10 ** 0.5)
is_prime_ex10 = True
for i in range(2, end_ex10 + 1):
    if num_ex10 % i == 0:
        is_prime_ex10 = False
        break
if is_prime_ex10:
    print(f'{num_ex10}是素数')
else:
    print(f'{num_ex10}不是素数')
```

### 练习 11 · 最大公约数（枚举法）
**Version:** 1.0 · **知识点:** for 循环、break

**演示数据:** `a=12, b=8` · **预期输出:** `a和b的最大公约数为4`


```python
a_ex11, b_ex11 = 12, 8
small_ex11 = min(a_ex11, b_ex11)
for i in range(1, small_ex11 + 1):
    if a_ex11 % i == 0 and b_ex11 % i == 0:
        print(f'a和b的最大公约数为{i}')
        break
```

### 练习 12 · 最大公约数（辗转相除法）
**Version:** 1.1 · **知识点:** while 循环

**演示数据:** `a=12, b=8` · **预期输出:** `最大公约数为4`


```python
a_ex12, b_ex12 = 12, 8
while a_ex12 % b_ex12 != 0:
    a_ex12, b_ex12 = b_ex12, a_ex12 % b_ex12
print(f'最大公约数为{b_ex12}')
```

### 练习 13 · 猜数字小游戏
**Version:** 1.0 · **知识点:** random、while 循环

> 需交互运行：单独执行此 cell，按提示输入猜测数字。


```python
import random

answer_ex13 = random.randint(1, 100)
count_ex13 = 0
print('猜字游戏开始‼️')
while True:
    guess_ex13 = int(input("请输入数字："))
    count_ex13 += 1
    if answer_ex13 > guess_ex13:
        print('猜小了')
    elif answer_ex13 < guess_ex13:
        print('猜大了')
    else:
        print('恭喜猜中了🎉')
        break
print(f'您一共猜了{count_ex13}次')
```
