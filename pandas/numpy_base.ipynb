#### 1.2随机数组

```python
# uniform() - 均匀分布的随机数组
np.random.uniform(-1,2,3) # 一维
np.random.uniform(-1,2,(2,3)) # 二维

# 特殊：rand() - 服从U[0,1]的均匀分布的数组
np.random.rand(3) # 一维
np.random.rand(3,3) # 二维
```

```python
# normal() - 生成N[μ，σ]正态分布的数组
mu,sigma = 3,2.5
np.random.normal(mu,sigma,3) # 一维
np.random.normal(mu,sigma,(3,3)) # 二维

# 特殊：randn() - 生成服从N[0，1]标准正态分布的数组
np.random.randn(3) # 一维
np.random.randn(3,3) # 二维
```

```python
# randint() - 生成在给定整数范围内的呈离散均匀分布的数组，high对应的数值不包括在内。
low,high,size = 1,10,(2,3)
np.random.randint(low,high,size)
```

```python
# choice() - 以一定概率和方式抽取元素，当不指定概率时为等概率抽样
my_list = ['a','b','c','d']
np.random.choice(
    my_list,
    3,
    replace= False, # 不放回
    p=[0.1,0.7,0.1,0.1] # 指定概率
)
np.random.choice(my_list,(3,3)) # 此时不能指定replace = False

#当返回的元素个数与原列表的元素个数相同时，不放回抽样等价于permutation()，即打散原列表。
np.random.permutation(my_list)
```

```python
# 随机种子 ->固定随机数的输出结果
np.random.seed(0)
np.random.rand() # 结果固定为0.5488135039273248
```

## 1-2NumPy数组的变形

### 1.由元素组织方式变化导致的变形

```python
# 维度交换方法：transpose()和T操作
carbon = np.random.rand(360,180,3)
res = carbon.transpose(2,0,1) # 将最后一个维度前置
res.shape
```

```python
# T操作
res = carbon.T # 特殊的transpose，将元素逆向变换
res.shape
# 特殊：矩阵转置 (行变列，列变行)
my_matrix = np.array([[2,3],[4,5]])
my_matrix.T
```

```python
# 重构维度：reshape()
my_matrix02 = np.arange(8)
my_matrix02.reshape(2,4,order="C") # 按行为顺序进行重构
my_matrix02.reshape(2,4,order="F") # 按列为顺序进行重构

#改变读取顺序
my_matrix02.reshape(2,4,order="C").reshape(8,order="F") # 按列为顺序进行读取

# -1占位：reshape()允许有一个维度存在空缺，只需填充-1。
my_matrix02.reshape(2,2,-1)
my_matrix02.reshape(2,-1,2)
my_matrix02.reshape(-1,2,2)
```

```python
# 特殊：多维度进行插入或者挤压维度,并且维度数为1
# expand_dims()插入维度
target = np.ones((2,2)) # 创建2x2的数组
expand_target = np.expand_dims(target,(0,2))  # 变形为1×2×1×2
expand_target.shape

# squeeze()挤压数组
np.squeeze(expand_target,(0,2))
```

### 2.由合并或拆分导致的变形

```python
# 合并：concatenate() & stack() - 在某一个维度上连接多个数组
# 创建一个具有实际含义的数组，维度为20×6××3，含义为年份（1981-2000），月份（1月-6月）与人口指标（出生率、死亡率和总人口数）

# stack()拼接后会产生新的维度
pop_man = np.random.rand(20,6,3) # 男性
pop_women = np.random.rand(20,6,3) # 女性
res = np.stack([pop_man,pop_women],axis=2)
res.shape
```

```python
# concatenate()是一种同质性或连续性拼接
pop_1_6 = np.random.rand(20,6,3) # 1-6月
pop_7_12 = np.random.rand(20,6,3) # 7-12月
res = np.concatenate([pop_1_6,pop_7_12],axis=1) # 将月份扩展到12个月
res.shape
```

```python
# 将两组月份割裂看待，而不是连续
res = np.stack([pop_1_6,pop_7_12],axis=1) # 表示在月份前加入上半年和下半年的维度
res.shape
```

```python
# 分割函数：split()
# indices_or_sections为整数则表明对某个维度的维数均匀分割
res = np.split(pop_1_6,indices_or_sections=3,axis=1)
for i in res:
    print(i.shape)
```

```python
# 分割函数：split()
# indices_or_sections为数组则表明对某个维度的维数分割点数 （不均匀分割）
res = np.split(pop_1_6,indices_or_sections=[1,4],axis=1)
for i in res:
    print(i.shape)
```

```python
# 练一练：split()函数从功能上更类似于concatenate()的逆操作还是stack()的逆操作？请说明理由。
# 更像concatenate()的逆操作。理由如下：concatenate不会产生新维度
arr = np.random.rand(10,20,30)
new_arr = np.concatenate(np.split(arr,indices_or_sections=5,axis=1),axis=1)
(arr == new_arr).all()
```
