```python
#练一练 1-4：用zip实现enumerate()函数
L2 = ["apple","ball","cat","dog","eye"]
L3 = list(i for i in range(len(L2)))
res = zip(L3,L2)
for index,value in res:
    print(index,value)

#优化
for index,value in zip(range(len(L2)),L2):
    print(index,value)
```
